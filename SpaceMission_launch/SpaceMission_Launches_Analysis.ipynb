# Conclusion

* The organization RVSN USSR conducted the highest number of space missions, with 1777 launches, followed by Arianespace and CASC.
* The most frequent launch location was Site 31/6, Baikonur Cosmodrome, Kazakhstan, where the USSR conducted 235 missions.
* The majority of missions resulted in success, with a success rate of approximately 89.7%, while failures accounted for around 7.8% of missions.
* USSR had the highest number of successful missions, with 1614 launches, indicating its dominance in the space exploration arena.
* Currently, out of the total rockets launched, only 790 satellites are active, while the status of 3534 satellites is retired.
* In conclusion, the analysis highlights the significant contributions of organizations like RVSN USSR and Arianespace to space exploration, with a high success rate overall despite some failures, and emphasizes the enduring impact of past missions through the large number of retired satellites.
