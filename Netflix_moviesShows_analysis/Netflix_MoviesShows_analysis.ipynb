# Netflix Movies and TV Shows Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\akash\Documents\dataset\kaggle dataset\Netflix movies,tv_shows.zip")
df
```

```python
df.info()
```

```python
df['director'].fillna('', inplace=True)
df['cast'].fillna('', inplace=True)
df['country'].fillna('', inplace=True)
df['date_added'].fillna('', inplace=True)
df['rating'].fillna('',inplace=True)
```

```python
df.dropna(subset=['duration'],inplace=True)
```

```python
df['type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Netflix shows')
plt.show()
```

- most of the Netflix shows are Movies.

```python
df['country'].value_counts()
```

```python
df['country'].value_counts().head(50).plot.barh(figsize=(12,7))
plt.title('Netflix Movies and TV Shows with Country')
plt.xlabel('Count')
```

```python
total=df['country'].value_counts().sum()
us = df['country'].value_counts().get('United States')
ind = df['country'].value_counts().get('India')
prcnt_ind = (ind/total)*100
prcnt_us = (us/total)*100
print(f'india {prcnt_ind} %')
print(f'US {prcnt_us} %')
```

- United States release more movies/tv shows in netflix.They release 2818 movies.
- India is in the second place with 972 movies/tv shows.
- 31.99% of movies/tv shows in netflix are from United states while 11.04% of movies/tv shows in netflix are from india

```python
df['release_year'].value_counts()
```

```python
yearly_type_counts = df.groupby(['release_year', 'type']).size().unstack()
```

```python
yearly_type_counts.plot(kind='bar', stacked=True, figsize=(15, 6))
plt.title('Distribution of Movies/Tv_shows Year-wise')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.legend(title='Type')
plt.show()
```

- Most of the Netflix shows are released(1147) in 2018.
- till 2018 the number of movies/tv shows releasing increases by year.And after 2018 the count of movie/tv shows releasing is decreasing.

# Conclusion

- The majority of content on Netflix comprises movies (69.6%) compared to TV shows (30.4%).
- The United States contributes the most content, with 31.99% of the total, followed by India at 11.04%.
- Most of the Netflix content released came out in 2018, with a declining trend in releases after that year.
- The most common rating for movies and TV shows on Netflix is TV-MA, followed by TV-14.
- The movie with the longest duration is "Black Mirror: Bandersnatch" at 312 minutes.
- The most common duration for movies and TV shows is 1 season, followed by 2 seasons.
- The most common genres of Netflix content are Dramas and International Movies, with Documentaries being the next most common.
- The most recently added movie is "Dick Johnson Is Dead".
