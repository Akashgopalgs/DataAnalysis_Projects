# Top 10 Highest-Paid Athletes(2011-2021) Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\akash\Documents\dataset\kaggle dataset\Highest-Paid Athletes 2011 - 2021\Athletes Earnings.csv")
df
```

```python
df.info()
```

```python
df['Prize Money'].fillna(df['Prize Money'].median(),inplace=True)
df['Endorsments'].fillna(0,inplace=True)
```

```python
df['Total Earnings'].sort_values(ascending=False).head(10)
```

```python
Highest_paidAthelets =df.sort_values(by='Total Earnings', ascending=False).head(10)
Highest_paidAthelets
```

```python
df_filtered = df.dropna(subset=['Total Earnings'])

plt.figure(figsize=(20, 6))
plt.bar(df_filtered['Player'], df_filtered['Total Earnings'], color='skyblue')
plt.xlabel('Player')
plt.ylabel('Total Earnings')
plt.title('Total Earnings of Players')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

- Here is the top 10 highest paid athelets (2011-2021 year).
- Lionel Messi is the highest paid athelet in the decade with 127M $.And he is also in second place
- Cristiano Ronaldo is in third place.
