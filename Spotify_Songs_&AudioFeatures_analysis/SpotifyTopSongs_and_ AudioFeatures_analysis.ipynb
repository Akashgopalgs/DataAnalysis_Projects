# Conclusion

- **Top Artists**: Taylor Swift has the most songs in the Spotify top songs list, followed by Drake. This indicates their popularity and consistent presence in the music charts.

- **Most Streamed Song**: "Blinding Lights" by The Weeknd is the most streamed song in the dataset, with over 3.5 billion streams. This suggests its widespread popularity and appeal among listeners.

- **Time Signature**: The majority of songs have a time signature of 4 beats, indicating a preference for common time in popular music.

- **Record Labels**: Columbia is the record label with the highest number of songs in the dataset, followed by Republic Records. This might reflect their influence and success in the music industry.

- **Song Duration**: Most songs in the dataset have durations ranging from 2.5 to 4.5 minutes, with the most common durations being around 3.3 to 3.4 minutes.

- **Energy and Loudness**: There is a positive correlation between energy and loudness, indicating that energetic songs tend to be louder. Most songs have energy levels between 0.5 to 0.8, and loudness levels are typically between -8 to -4 dB.

- **Danceability**: "Botadinha Saliente" by MC Rogerinho is the most danceable song in the dataset, with a high number of streams. Generally, danceability scores for songs range from 0.65 to 0.85.
