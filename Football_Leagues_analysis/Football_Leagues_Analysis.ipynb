# Conclusion

* Cristiano Ronaldo and Lionel Messi are two of the greatest players of all time. Messi has consistently outperformed Ronaldo in terms of matches played and goals scored over the last five years.
* Italian Serie A had the most total goals scored among top football leagues from 2016 to 2020, with players scoring 1359 goals during this period.
* La Liga had the highest average goals per match, followed by the Bundesliga, indicating a more attacking style of play in these leagues.
* Serie A recorded the highest number of shots during matches among top leagues, demonstrating an emphasis on offensive play.
* The top leagues in the world, namely the Premier League, La Liga, Serie A, and Bundesliga, exhibit strong correlations between goals and expected goals (xG), shots on target, and total shots, indicating that efficient attacking play leads to more goals.
* Matches played and minutes played show a positive but relatively weak correlation with goals, suggesting that players who participate more frequently tend to score more goals, albeit with less influence compared to factors like shots and on-target shots.
* In summary, efficient attacking play, as indicated by high xG, shots on target, and total shots, significantly contributes to goal-scoring success in top football leagues. Additionally, player performance, particularly in terms of matches played, minutes played, and shots, has a notable impact on goal-scoring outcomes.
